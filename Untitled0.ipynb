```python
import requests
import pandas as pd
import time
import csv
import matplotlib.pyplot as plt

# Налаштування
dates = pd.date_range(start="2025-01-01", end="2025-01-22")
total_btc = 0
total_invested = 0  # Загальна сума вкладень
btc_data_list = []  # Для збереження даних
portfolio_values = []  # Для вартості портфеля

# Отримання даних через API
print("Отримуємо дані про курс BTC...\n")
for i, date in enumerate(dates):
    formatted_date = date.strftime("%d-%m-%Y")
    url = f"https://api.coingecko.com/api/v3/coins/bitcoin/history?date={formatted_date}&localization=false"
    response = requests.get(url)

    if response.status_code == 200:
        btc_data = response.json()
        if "market_data" in btc_data:
            btc_price = btc_data['market_data']['current_price']['usd']
            daily_btc = 10 / btc_price
            total_btc += daily_btc
            total_invested += 10
            portfolio_value = total_btc * btc_price  # Вартість портфеля на поточний день
            btc_data_list.append([formatted_date, f"{daily_btc:.8f}", f"{btc_price:.2f}", total_invested, f"{total_btc:.8f}", f"{portfolio_value:.2f}"])
            portfolio_values.append(portfolio_value)
            print(f"{formatted_date}: Куплено {daily_btc:.8f} BTC за 10 USD (курс {btc_price:.2f} USD, "
                  f"загальна інвестиція: {total_invested} USD, загальний BTC: {total_btc:.8f}, "
                  f"вартість портфеля: {portfolio_value:.2f} USD)")
        else:
            btc_data_list.append([formatted_date, "Дані відсутні", "Дані відсутні", total_invested, f"{total_btc:.8f}", "Дані відсутні"])
            portfolio_values.append(None)
            print(f"Дані відсутні за {formatted_date}")
    else:
        btc_data_list.append([formatted_date, "Помилка API", "Помилка API", total_invested, f"{total_btc:.8f}", "Помилка API"])
        portfolio_values.append(None)
        print(f"Не вдалося отримати дані за {formatted_date} (Статус: {response.status_code})")

    # Обмеження: не більше 8 запитів за хвилину
    if (i + 1) % 8 == 0:  # Кожні 8 запитів чекаємо 30 секунд
        print("Досягнуто ліміту запитів. Очікування 30 секунд...")
        time.sleep(36)
    else:
        time.sleep(11)  # Затримка між запитами

# Підсумок
print(f"\nСумарна кількість BTC за 22 дні: {total_btc:.8f}")
print(f"Загальна сума вкладень: {total_invested} USD")

# Збереження даних у CSV
csv_filename = "btc_portfolio.csv"
print(f"\nЗберігаємо результати у файл '{csv_filename}'...")
with open(csv_filename, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Дата", "Куплено BTC", "Курс (USD)", "Загальна сума вкладень (USD)", "Загальний BTC", "Вартість портфеля (USD)"])
    writer.writerows(btc_data_list)
print(f"Дані збережено у файл '{csv_filename}'")

# Візуалізація даних
print("\nБудуємо графік...")

# Дані для графіка
dates_for_plot = [row[0] for row in btc_data_list if row[1] != "Дані відсутні" and row[1] != "Помилка API"]
prices_for_plot = [float(row[2]) for row in btc_data_list if row[1] != "Дані відсутні" and row[1] != "Помилка API"]
portfolio_for_plot = [float(row[5]) for row in btc_data_list if row[1] != "Дані відсутні" and row[1] != "Помилка API"]
invested_for_plot = [row[3] for row in btc_data_list if row[1] != "Дані відсутні" and row[1] != "Помилка API"]

# Створення графіка
fig, ax1 = plt.subplots(figsize=(12, 6))

# Перша вісь Y - курс BTC
ax1.plot(dates_for_plot, prices_for_plot, 'b-o', label="Курс BTC (USD)")
ax1.set_xlabel("Дата")
ax1.set_ylabel("Курс BTC (USD)", color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.grid()

# Друга вісь Y - вартість портфеля та загальна сума вкладень
ax2 = ax1.twinx()
ax2.plot(dates_for_plot, portfolio_for_plot, 'g-o', label="Вартість портфеля (USD)")
ax2.plot(dates_for_plot, invested_for_plot, 'y--', label="Загальна сума вкладень (USD)")
ax2.set_ylabel("Вартість (USD)", color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Легенда
fig.tight_layout()
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.title("Динаміка курсу BTC, вартості портфеля та загальної суми вкладень (01-22 січня 2025)")
plt.show()
```
